# Create Torch Dataset

```python
class KeypointsDataset(Dataset):
    def __init__(self,img_dir,data_file):
        self.img_dir = img_dir
        with open(data_file, "r") as f:
            self.data = json.load(f)
        #   standardize and normalise the images to same size    
        #  transformation pipeline
        self.transforms = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((224,224)),
            transforms.ToTensor(),
            transforms.Normalize(mean= [0.485, 0.456 , 0.406], std = [0.229, 0.224 , 0.225])
        ])
        #specify length of the dataset 
    def __len__(self):
        return len(self.data)
        # getting items batch by batch and train it, takes an index and returns the image and  returns the keypoint,
    def __getitem__(self, idx):
        item = self.data[idx]
        img = cv2.imread(f"{self.img_dir}/{item['id']}.png")
        h ,w = img.shape[:2]
        img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
        img = self.transforms(img)
        kps = np.array(item['kps']).flatten()
        kps = kps.astype(np.float32)
        #  the images resized, we have to match the keypoints accordingly.
        
        kps[::2] *= 224.0/w
        kps[1::2] *= 224.0/h
        
        return img,kps
        
```

```python
train_dataset = KeypointsDataset("data/images","data/data_train.json")
val_dataset = KeypointsDataset("data/images","data/data_val.json")
# loads the data in batches 
train_loader = DataLoader(train_dataset , batch_size=8 , shuffle = True)
val_loader = DataLoader(val_dataset , batch_size=8 , shuffle = True)
```

# Model 

```python
model = models.resnet50(pretrained = True)
# Downloads Resnet50 , a pretrained CNN. We replace the final layer here in our problem statement for keypoint detectors.

model.fc = torch.nn.Linear(model.fc.in_features, 14*2) # replaces the last layer of the neural network.
```

```python
model = model.to(device)
```

```python
criterion = torch.nn.MSEloss()
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-4)
```

# Training the model 

```python
epochs = 20 
for epoch in range(epochs):
    for i , (imgs,kps) in enumerate(train_loader):
        imgs = imgs.to(device)
        kps = kps.to(device)
        
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion (outputs , kps)
        loss.backward()
        optimizer.step()
        
        if i % 10 == 0 :
            print(f"Epoch {epoch}, iter {i} , loss {loss.item()}")
        
```

```python
torch.save(model.state_dict(), "keypoints_model.pth")
```
